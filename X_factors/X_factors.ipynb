```python
import jaqs_fxdayu
jaqs_fxdayu.patch_all()
from jaqs.data import DataView
from jaqs.data import RemoteDataService
from jaqs_fxdayu.data.dataservice import LocalDataService
import os
import numpy as np
import pandas as pd
import warnings

warnings.filterwarnings("ignore")

dataview_folder = '../data'
dv = DataView()
ds = LocalDataService(fp=dataview_folder)


factor_list = ['volume']
check_factor = ','.join(factor_list)

dv_props = {'start_date': start, 'end_date': end, 'symbol':','.join(stock_symbol),
         'fields': check_factor,
         'freq': 1,
         "prepare_fields": True}

dv.init_from_config(dv_props, data_api=ds)
dv.prepare_data()
```

## BM 账面市值比 = 股东权益（净资产）/市值

```python
dv.add_field('tot_assets', ds)
dv.add_field('total_liab', ds)
dv.add_field('total_mv', ds)

dv.add_formula('BM', '(tot_assets-total_liab)/total_mv',
              is_quarterly=False, add_data=True)

# IC=0.06, IR=0.41
```

## 高低均价与成交量
参照alpha13

```python
dv.add_formula('HLV', '((high*low*vwap)^(1/3) * volume)',
              is_quarterly=False, add_data=True)

# IC=-0.03， IR=-0.27
```

## 220日净利率变动

```python
# 收入净利率占比变动
dv.add_field('net_profit', ds)

dv.add_formula('NP_220', 
               '(net_profit-Delay(net_profit, {}))/Delay(net_profit, {})'.format(220, 220),
               is_quarterly=False, add_data=True)

# IC=0.13， IR=0.25
```

## 60日情绪指标AR

```python
dv.add_formula('AR_60',
              'Ts_Sum(high-open, {})/Ts_Sum(open-low, {})'.format(60, 60),
              is_quarterly=False, add_data=True)

# IC=-0.03, IR=-0.31
```

## 60日交易量与收盘价
描述：60日前的交易量比上收盘价除以今日交易量比上收盘价

```python
dv.add_formula('VDC_60',
           '(volume/Delay(close, {}))/(volume/close)'.format(60),
          is_quarterly=False, add_data=True)
```

## 换手率
10日换手率比120日换手率乘以20日换手率比120日换手率

```python
# 10日换手率-
dv.add_field('VOL10', ds)
dv.add_field('VOL20', ds)
dv.add_field('VOL120', ds)
dv.add_field('VOL240', ds)

dv.add_formula('VOL_R',
              '(VOL10/VOL120)*(VOL20/VOL120)', 
               is_quarterly=False, add_data=True)

# IC=-0.04, IR=-0.46
```

10日换手率乘20日换手率乘120日换手率乘240日换手率的立方根

```python
dv.add_formula('VOL_T',
              '(VOL10*VOL20*VOL240*VOL120)^(1/4)',
               is_quarterly=False, add_data=True)

# IC=-0.10, IR=-0.77
```

## 多头力道与空头力道

```python
dv.add_field('BearPower', ds)
dv.add_field('BullPower', ds)
dv.add_formula('BB',
              'BearPower-BullPower', 
               is_quarterly=False, add_data=True)

# IC=0.03, IR=0.30
```

```python
# 20日相关系数
dv.add_formula('BBR_20',
              'Correlation(BearPower, BullPower, {})'.format(20),
              is_quarterly=False, add_data=True)

# IC=0.05, IR=0.66
```

## 90日前交易量比收盘价除以今日数据后的90日标准差

```python
# SDVC_90
dv.add_formula('SDVC_90',
                   'StdDev(Delay(volume/close, {})/(volume/close), {})'.format(1, 90),
                  is_quarterly=False, add_data=True)
```

## 120日前开收盘价乘积的平方除今日数据后的120日标准差

```python
dv.add_formula('SDOC_120',
                   'StdDev(Delay((open*close)^0.5, {})/((open*close)^0.5), {})'.format(120, 120),
                   is_quarterly=False, add_data=True)
```
